DATA CLEANING

```python
condition = df_fsr['Heard_before'] =='No'
new_text ='No'
df_fsr['Tried_before'] = np.where(condition,new_text,df_fsr['Tried_before'])
df_fsr
```

```python
condition = df_fsr['Tried_before'] =='Yes'
new_text ='Already tried'
df_fsr['Reasons_preventing_trying'] = np.where(condition,new_text,df_fsr['Reasons_preventing_trying'])
df_fsr
```

```python
condition = df_fsr['Heard_before'] =='No'
new_value =''
df_fsr['Brand_perception'] = np.where(condition,new_value,df_fsr['Brand_perception'])
df_fsr
```

1. DEMOGRAPHIC INSIGHTS:

a. Who prefers energy drink more?

```python
df_selected = df_respondents[['Respondent_ID','Age', 'Gender']]
df_fsr_new = pd.merge(df_fsr,df_selected, on = 'Respondent_ID', how = 'left')
df_fsr_new
```

a. Who prefers energy drink more?

```python
Gender = df_fsr_new['Gender'].value_counts(normalize = True)*100
print('Prefered Gender:')
for index, percentage in Gender.items():
    print(f'{index}: {percentage}%')
explode =(0.05,0,0)
colors =['#007ea7','#80ced7','#ccdbdc']
plt.figure(facecolor='#e2eafc')
plt.pie(Gender, labels = Gender.index, colors = colors, autopct='%1.1f%%', startangle= 90, shadow =True, explode = explode)
plt.title('Preference for energy drink')
```

b. Which age group prefers energy drink more?

```python
Age = df_fsr_new['Age'].value_counts()
print('Prefered age group:')
for value, count in Age.items():
    print(f'{value}: {count}')

fig, ax = plt.subplots()
Age.plot(kind='bar', color ='#1b4332',ax=ax)
ax.set_facecolor('#d8f3dc')
plt.title('Age group who prefers energy drink:')
plt.ylabel('Count')
plt.xticks(rotation=0)
for i, v in enumerate(Age):
    fig, ax.text(i, v + 0.1, str(v), ha='center', va='bottom')
```

5. BRAND PENETRATION:

b. Which cities do we need to focus more on?

```python
selected = df_respondents[['Respondent_ID','City_ID']]
df_fsr_5 = pd.merge(df_fsr,selected, on = 'Respondent_ID', how = 'left')
selected_b = df_cities[['City_ID','City']]
df_fsr_5b = pd.merge(df_fsr_5, selected_b, on = 'City_ID', how = 'left')

negative_response = (df_fsr_5b[df_fsr_5b['Brand_perception']=='Negative'])
city_with_negative_response = negative_response['City'].value_counts()
city_count = df_fsr_5b['City'].value_counts()
negative_response_per_city = (round(city_with_negative_response*100/city_count,2)).sort_values(ascending = False)
print('The cities with highest negative response to Codex brand are:')
for city, negative_response in negative_response_per_city.items():
    print(f'{city}: {negative_response}%')


fig, ax = plt.subplots(figsize=(10,6))
negative_response_per_city.plot(kind='bar', color ='#003249',ax=ax)
ax.set_facecolor('#d8f3dc')
plt.xlabel('City')
plt.ylabel('Percentage(%)')
plt.title('The cities with highest negative response to Codex brand are:')
plt.xticks(rotation=45)
for i, v in enumerate(negative_response_per_city):
    fig, ax.text(i, v + 0.1, str(v), ha='center', va='bottom')
```

```python
▫️ The above mentioned are the cities with respondents who shared negative perception about our brand.
▫️ So we need to focus on the cities having higher negative responses and has to make them positive for growing our brand.
```
